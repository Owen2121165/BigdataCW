## 1: Prime Number Check

```python

# defines prime number
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# user input
num = int(input("Enter a number: "))
# checks for the input, output based on the input and responds accordingly
if is_prime(num):
    print("This is a prime number.")
else:
    print("This is not a prime number.")
```

## 2: Fibonacci Sequence

```python

# Define a function that generates the sequence to 'n' terms
def fibonacci(n):
    fib_sequence = [0, 1]
    # Loop from index 2 up to (n-1)
    for i in range(2, n): 
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
    # returns the first n
    return fib_sequence[:n]
# ask for how many Fibonacci numbers to be generated
n = int(input("Insert the amount of Fibonacci numbers to be generated: "))
print("Fibonacci Sequence:", fibonacci(n))
```

## 3: Triangle Pattern

```python

def print_pattern(rows):
    # defines how the shape is formed
    for i in range(rows, 0, -1):
    # prints 0 in a pattern of "i" from above
        print(" * " * i)
# prints in number of rows you needed
print_pattern(5)
```

## 5: Min & Max Finder

```python

# finds the largest and smallest value
def min_max(iterable):
    return min(iterable), max(iterable)

numbers = [3, 5, 7, 2, 8, 10, 4]
# assigning min and max seperately
min_val, max_val = min_max(numbers)
# prints the soulution
print("Min:", min_val)
print("Max:", max_val)
```

## 8: Mathematical Series Calculation

```python

#import the math lib
import math
# initiates to 0, which stores the sum of the series
def series_calculation(x, N):
    result = 0
# Loops from 1 to N
    for i in range(1, N + 1):
    # forumla
        result += (x * (5 ** i) ** 2) / (i + math.factorial(i + 1))
    return result
# user input
x = int(input("Enter x: "))
N = int(input("Enter N: "))
# prints the results
print("Series result:", series_calculation(x, N))
```

## Exercise 9: Image Reflection

```python

# imported lib
import numpy as np
import matplotlib.pyplot as plt
import PIL
import urllib

# import images from the internet
image_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/8/8c/1994_Mazda_RX-7_R2_in_Vintage_Red%2C_front_left_%28Lime_Rock%29.jpg/280px-1994_Mazda_RX-7_R2_in_Vintage_Red%2C_front_left_%28Lime_Rock%29.jpg"
image_data = urllib.request.urlopen(image_url)
image = PIL.Image.open(image_data)


# Converts image to NumPy array
np_image = np.array(image)

# Corrected variable names
horizontal_flip = np.fliplr(np_image)
vertical_flip = np.flipud(np_image)
both_flipped = np.flipud(np.fliplr(np_image))

# Display original and flipped images
plt.figure(figsize=(10, 7))

# Original Image
plt.subplot(2, 2, 1) # location of the image
plt.imshow(np_image)
plt.title("Original Image") # title of the image

# Horizontally Flipped
plt.subplot(2, 2, 2)
plt.imshow(horizontal_flip)
plt.title("Horizontally Flipped")

# Vertically Flipped
plt.subplot(2, 2, 3)
plt.imshow(vertical_flip)
plt.title("Vertically Flipped")

# Flipped Both Ways
plt.subplot(2, 2, 4)
plt.imshow(both_flipped)
plt.title("Both Flipped")

# Display all images
plt.tight_layout()
plt.show()
```

## 10: Temperature Data

```python
import numpy as np
# temp data
temps_celsius = np.array([22, 24, 19, 21, 23, 25, 26, 27, 24, 23, 20, 22, 21, 19, 18, 22, 23, 24, 26, 28, 30, 27, 25, 26, 24, 22, 21, 20, 19, 23])

# Calculations
# min temp
min_temp = np.min(temps_celsius)
# max temp
max_temp = np.max(temps_celsius)
# avg temp 
avg_temp = np.mean(temps_celsius)
# convert to fahrenheit
temps_fahrenheit = (temps_celsius * 9/5) + 32
# count for temp above 25
above_25 = temps_celsius[temps_celsius > 25]
# count for temp below 20
below_20_count = np.sum(temps_celsius < 20)
# sort in ascending order
sorted_temps = np.sort(temps_celsius)
# Reshape into 5 weeks of 6 days each
reshaped_temps = temps_celsius.reshape(5, 6)  

# Print results
print(f"Min Temperature: {min_temp}C")
print(f"Max Temperature: {max_temp}C")
print(f"Average Temperature: {avg_temp:.2f}C")
print(f"Temperatures in Fahrenheit: {temps_fahrenheit}")
print(f"Days with temperature > 25C: {above_25}")
print(f"Days with temperature < 20C: {below_20_count}")
print("Sorted Temperatures:", sorted_temps)
print("Reshaped Temperature Data:\n", reshaped_temps)
```
